# *****let's import dependencies*****

```python
import os
import shutil
import ultralytics
from ultralytics import YOLO
os.environ["WANDB_DISABLED"] = "true"
```

# **Start train**

```python
model = YOLO('yolov8n.pt')  # build a new model from scratch
results = model.train(data='/kaggle/input/bone-fracture-detection-computer-vision-project/data.yaml', epochs=100)  # train the model
```

# **Eval model**

```python
results = model.val()  # evaluate model performance on the validation set
```

# **Predict**

```python
results = model('/kaggle/input/bone-fracture-detection-computer-vision-project/test/images/distal-humerus-fracture-1_jpg.rf.831cb137cfcbde1079f86abd5f5f2867.jpg')  # predict on an image
```

# **Export**

```python
results = model.export(format='onnx')  # export the model to ONNX format
```
